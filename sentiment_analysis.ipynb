```
import nltk

nltk.download('vader_lexicon')
```

```
#Configure the sentiment analysis function
sentiment_analyzer = SentimentIntensityAnalyzer()

list_of_sentiments = []

for index, row in data.iterrows():
  tweet_polarity_scores = sentiment_analyzer.polarity_scores(row['Tweets'])
  list_of_sentiments.append(tweet_polarity_scores)
  
polarity_series = pd.Series(list_of_sentiments)
data['polarity'] = polarity_series.values

display(data.head(100))
  
 
```
